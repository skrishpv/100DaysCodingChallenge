## Scatterplot vs Hexagonal binning plot

```python
import pandas as pd
import numpy as np
```

Comparing two numerical variables is a common step in exploratory analysis. Though there are many ways to visualize this comparison the best way and the most frequently used way is to plot a scatterplot. 
A scatterplot gives us an indication of the relationship between the two variables and their correlation with each other.
Scatterplots ensure that every datapoint in the sample is plotted and represented in the graph. This can be both an advantage and a disadvantage.

While, it is great to see every datapoint separately this is only practical when the number of data points is relatively small. For data sets with hundreds of thousands or millions of records, a scatterplot would be too dense, so we need a different way to visualize the relationship between the variables being compared.

This is where a Hexagonal Bin Plot becomes really useful. Observations close to each other are put into hexagonal bins and the color of the bins indicate the number of records in that bin. This makes the visuals really intuitive.

```python
kc_tax = pd.read_csv('Data/kc_tax.csv.gz', compression='gzip')
```

```python
kc_tax.head()
```

```python
kc_tax_sub = kc_tax.loc[(kc_tax.TaxAssessedValue < 750000) &
                       (kc_tax.SqFtTotLiving >100)&
                       (kc_tax.SqFtTotLiving <3500),:]
```

```python
scatterplot = kc_tax_sub.plot.scatter(x='SqFtTotLiving', y='TaxAssessedValue', figsize=(4,4))
scatterplot.set_xlabel('Square Feet to Living')
scatterplot.set_ylabel('Tax Assessed Value')
scatterplot
```

```python
hexbinplot = kc_tax_sub.plot.hexbin(x='SqFtTotLiving', y= 'TaxAssessedValue',
                           gridsize=30, sharex=False, figsize=(4,4))
hexbinplot.set_xlabel('Square Feet to Living')
hexbinplot.set_ylabel('Tax Assessed Value')

hexbinplot
```
